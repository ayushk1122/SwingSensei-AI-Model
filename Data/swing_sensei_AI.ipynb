FPS calculation

```python
import numpy as np
import pandas as pd

def calculate_fps(frame_times_df):
    # Calculate the differences between consecutive frame times
    time_diffs = frame_times_df['timestamp'].diff().dropna()

    # Filter out any negative or zero time differences (if any exist due to data noise)
    time_diffs = time_diffs[time_diffs > 0]

    # Calculate FPS as the inverse of the average time difference
    avg_time_diff = time_diffs.mean()
    fps = 1.0 / avg_time_diff if avg_time_diff > 0 else 0

    return fps

# Example usage
frame_times_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/frame_times.csv')
calculated_fps = calculate_fps(frame_times_df)
print(f"Calculated FPS: {calculated_fps:.2f}")



```

Peak Hand/Swing Speed Calculation (working prototype)

```python
import numpy as np
import pandas as pd

def calculate_hand_speed_with_bbox(keypoints_df, fps, real_world_height_cm, slow_motion_factor=1, frame_interval = 3):
    hand_speeds = []

    # Calculate bounding box height in pixels
    frameBB = keypoints_df[keypoints_df['frame'] == 0]
    bbox_top_y = frameBB[frameBB['body_part'] == 'bounding_box_top']['y'].values[0]
    bbox_bottom_y = frameBB[frameBB['body_part'] == 'bounding_box_bottom']['y'].values[0]
    bbox_height_pixels = bbox_bottom_y - bbox_top_y

    # Calculate pixel-to-cm conversion factor
    pixel_to_cm_conversion = real_world_height_cm / bbox_height_pixels

    for frame in range(keypoints_df['frame'].min(), keypoints_df['frame'].max() - 3, frame_interval):
        frame1 = keypoints_df[keypoints_df['frame'] == frame]
        frame2 = keypoints_df[keypoints_df['frame'] == frame + frame_interval]

        if frame1.empty or frame2.empty:
            print(f"Skipping Frame {frame}: No data for frame1 or frame2.")
            continue

        # Extract wrist coordinates
        if frame1[frame1['body_part'] == 'left_wrist'].empty or frame1[frame1['body_part'] == 'right_wrist'].empty:
            print(f"Frame {frame}: Missing left or right wrist data.")
            continue

        if frame2[frame2['body_part'] == 'left_wrist'].empty or frame2[frame2['body_part'] == 'right_wrist'].empty:
            print(f"Frame {frame + 6}: Missing left or right wrist data.")
            continue

        left_hand1 = frame1[frame1['body_part'] == 'left_wrist'][['x', 'y']].values[0]
        right_hand1 = frame1[frame1['body_part'] == 'right_wrist'][['x', 'y']].values[0]
        left_hand2 = frame2[frame2['body_part'] == 'left_wrist'][['x', 'y']].values[0]
        right_hand2 = frame2[frame2['body_part'] == 'right_wrist'][['x', 'y']].values[0]

        # Extract bounding box height from keypoints data


        # Calculate hand movement distances in cm
        left_hand_distance_pixels = np.linalg.norm(left_hand2 - left_hand1)
        right_hand_distance_pixels = np.linalg.norm(right_hand2 - right_hand1)

        # Convert pixel distances to real-world distances in cm
        left_hand_distance_cm = left_hand_distance_pixels * pixel_to_cm_conversion
        right_hand_distance_cm = right_hand_distance_pixels * pixel_to_cm_conversion

        # Average hand speed in cm/s
        avg_hand_speed_cm_per_sec = ((left_hand_distance_cm + right_hand_distance_cm) / 2) * fps

        # Convert to mph
        speed_mph = (avg_hand_speed_cm_per_sec * 0.01 * 3600) / 1609.34  # Convert cm/s to mph

        # Adjust for real-time speed considering slow-motion
        real_time_speed_mph = speed_mph * slow_motion_factor

        hand_speeds.append(real_time_speed_mph)

        print(f"Frame {frame}: Avg Hand Speed in MPH: {real_time_speed_mph:.2f} mph")

    peak_hand_speed = max(hand_speeds) if hand_speeds else 0.0
    print(f"Peak Hand Speed: {peak_hand_speed:.2f} mph")
    return peak_hand_speed

# Example usage:
keypoints_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/keypoints.csv')

# Real-world height of Aaron Judge in cm
real_world_height_cm = 160.2  # For Aaron Judge, 6'7" is approximately 200 cm

frame_times_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/frame_times.csv')
vid_fps = calculate_fps(frame_times_df)
slow_motion_factor = 30 / vid_fps
# frame_interval = 3
# Calculate Peak Hand Speed
peak_hand_speed = calculate_hand_speed_with_bbox(keypoints_df, calculated_fps, real_world_height_cm, slow_motion_factor, frame_interval)
```

Weight distribution during swing calc (working prototype)

```python
import numpy as np
import pandas as pd

def calculate_weight_distribution_with_bbox(keypoints_df, real_world_height_cm, frame_interval):
    weight_distributions = []

    # Calculate bounding box height in pixels from the first frame
    frameBB = keypoints_df[keypoints_df['frame'] == keypoints_df['frame'].min()]
    bbox_top_y = frameBB[frameBB['body_part'] == 'bounding_box_top']['y'].values[0]
    bbox_bottom_y = frameBB[frameBB['body_part'] == 'bounding_box_bottom']['y'].values[0]
    bbox_height_pixels = bbox_bottom_y - bbox_top_y

    # Calculate pixel-to-cm conversion factor
    pixel_to_cm_conversion = real_world_height_cm / bbox_height_pixels

    for frame in range(keypoints_df['frame'].min(), keypoints_df['frame'].max() + frame_interval):
        frame_data = keypoints_df[keypoints_df['frame'] == frame]

        # Check if the necessary keypoints exist
        if not frame_data[frame_data['body_part'] == 'left_ankle'].empty and \
           not frame_data[frame_data['body_part'] == 'right_ankle'].empty and \
           not frame_data[frame_data['body_part'] == 'left_hip'].empty and \
           not frame_data[frame_data['body_part'] == 'right_hip'].empty:

            left_ankle = frame_data[frame_data['body_part'] == 'left_ankle'][['x', 'y']].values[0]
            right_ankle = frame_data[frame_data['body_part'] == 'right_ankle'][['x', 'y']].values[0]
            left_hip = frame_data[frame_data['body_part'] == 'left_hip'][['x', 'y']].values[0]
            right_hip = frame_data[frame_data['body_part'] == 'right_hip'][['x', 'y']].values[0]

            # Convert pixel distances to real-world distances (cm)
            left_ankle_real = left_ankle * pixel_to_cm_conversion
            right_ankle_real = right_ankle * pixel_to_cm_conversion
            left_hip_real = left_hip * pixel_to_cm_conversion
            right_hip_real = right_hip * pixel_to_cm_conversion

            # Calculate the center of mass as the midpoint between the hips
            CoM = np.mean([left_hip_real, right_hip_real], axis=0)

            # Calculate weight distribution in real-world coordinates
            left_foot_weight = (right_ankle_real[0] - CoM[0]) / (right_ankle_real[0] - left_ankle_real[0]) * 100
            right_foot_weight = 100 - left_foot_weight

            weight_distributions.append((frame, left_foot_weight, right_foot_weight))
        else:
            continue

    return weight_distributions

# Example usage
keypoints_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/keypoints.csv')
real_world_height_cm = 160.2  # Example player height in cm

# frame_interval = 3
weight_distribution = calculate_weight_distribution_with_bbox(keypoints_df, real_world_height_cm, frame_interval)

# Print out the weight distribution for each frame
for frame, left_weight, right_weight in weight_distribution:
    print(f"Frame {frame}: Left Foot: {left_weight:.2f}%, Right Foot: {right_weight:.2f}%")
```

Head Movement Calculation

```python
import numpy as np
import pandas as pd

def calculate_head_stability_in_swing_phase(keypoints_df, real_world_height_cm, frame_interval):
    # Conversion factor from cm to inches
    cm_to_inch = 0.393701

    # Calculate bounding box height in pixels from the first frame
    frameBB = keypoints_df[keypoints_df['frame'] == keypoints_df['frame'].min()]
    bbox_top_y = frameBB[frameBB['body_part'] == 'bounding_box_top']['y'].values[0]
    bbox_bottom_y = frameBB[frameBB['body_part'] == 'bounding_box_bottom']['y'].values[0]
    bbox_height_pixels = bbox_bottom_y - bbox_top_y

    # Calculate pixel-to-cm conversion factor
    pixel_to_cm_conversion = real_world_height_cm / bbox_height_pixels

    # Split the video into 3 phases
    total_frames = keypoints_df['frame'].max() - keypoints_df['frame'].min() + 1
    frames_per_phase = total_frames // 3  # Number of frames in each phase

    # We are only interested in the 3rd phase (swing phase)
    start_frame = keypoints_df['frame'].min() + 2 * frames_per_phase
    end_frame = keypoints_df['frame'].max()

    # Initialize variables for horizontal head movement calculation
    horizontal_head_movement_cm = 0.0
    previous_nose_x_position = None

    # Print out details about phase 3 frames being used for the calculation
    print(f"Phase 3 (Swing Phase) begins at frame {start_frame} and ends at frame {end_frame}.")
    print(f"Processing every {frame_interval} frames.\n")

    # Iterate through frames in the 3rd phase (swing phase) with the specified frame interval
    for frame in range(start_frame, end_frame + 1, frame_interval):
        frame_data = keypoints_df[keypoints_df['frame'] == frame]

        # Check if nose keypoint exists (as a proxy for head)
        if not frame_data[frame_data['body_part'] == 'nose'].empty:
            nose_position = frame_data[frame_data['body_part'] == 'nose'][['x', 'y']].values[0]
            nose_x_position_real = nose_position[0] * pixel_to_cm_conversion  # Only track horizontal (x-axis) movement

            # Print the frame number and nose position
            print(f"Frame {frame}: Nose X-Position (in pixels): {nose_position[0]}, Real-world X-Position: {nose_x_position_real:.2f} cm")

            # If this is not the first frame, calculate the horizontal distance moved by the nose
            if previous_nose_x_position is not None:
                movement_distance_x = abs(nose_x_position_real - previous_nose_x_position)
                horizontal_head_movement_cm += movement_distance_x

            # Update the previous nose x position
            previous_nose_x_position = nose_x_position_real
        else:
            print(f"Frame {frame}: Missing nose keypoint data. Skipping this frame.")

    # Convert total horizontal head movement from cm to inches
    horizontal_head_movement_in = horizontal_head_movement_cm * cm_to_inch

    print(f"\nTotal Horizontal Head Movement in Swing Phase: {horizontal_head_movement_in:.2f} inches")
    return horizontal_head_movement_in

# Example usage:
keypoints_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/keypoints.csv')
real_world_height_cm = 160.2  # Example player height in cm
# frame_interval = 3  # Use the specified frame interval

# Calculate Head Stability in the Swing Phase (Phase 3) with bounding box calibration
total_horizontal_head_movement = calculate_head_stability_in_swing_phase(keypoints_df, real_world_height_cm, frame_interval)
```

Extension/follow through Calculation

```python
import numpy as np
import pandas as pd

def calculate_angle(a, b, c):
    """ Calculate angle using three points (a, b, c) where b is the vertex. """
    ab = np.linalg.norm(a - b)
    bc = np.linalg.norm(b - c)
    ac = np.linalg.norm(a - c)
    if ab == 0 or bc == 0:  # Prevent division by zero
        return 0
    return np.degrees(np.arccos((ab**2 + bc**2 - ac**2) / (2 * ab * bc)))

def get_pose_angles(frame_data):
    """ Calculate key angles for the pose in the given frame. """
    # Extract key body parts for calculating angles
    left_shoulder = frame_data[frame_data['body_part'] == 'left_shoulder'][['x', 'y']].values[0]
    left_elbow = frame_data[frame_data['body_part'] == 'left_elbow'][['x', 'y']].values[0]
    left_wrist = frame_data[frame_data['body_part'] == 'left_wrist'][['x', 'y']].values[0]
    right_shoulder = frame_data[frame_data['body_part'] == 'right_shoulder'][['x', 'y']].values[0]
    right_elbow = frame_data[frame_data['body_part'] == 'right_elbow'][['x', 'y']].values[0]
    right_wrist = frame_data[frame_data['body_part'] == 'right_wrist'][['x', 'y']].values[0]
    left_hip = frame_data[frame_data['body_part'] == 'left_hip'][['x', 'y']].values[0]
    right_hip = frame_data[frame_data['body_part'] == 'right_hip'][['x', 'y']].values[0]
    left_knee = frame_data[frame_data['body_part'] == 'left_knee'][['x', 'y']].values[0]
    right_knee = frame_data[frame_data['body_part'] == 'right_knee'][['x', 'y']].values[0]
    left_ankle = frame_data[frame_data['body_part'] == 'left_ankle'][['x', 'y']].values[0]
    right_ankle = frame_data[frame_data['body_part'] == 'right_ankle'][['x', 'y']].values[0]

    # Calculate angles
    left_arm_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
    right_arm_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
    left_leg_angle = calculate_angle(left_hip, left_knee, left_ankle)
    right_leg_angle = calculate_angle(right_hip, right_knee, right_ankle)
    torso_angle = calculate_angle(left_shoulder, left_hip, left_knee)  # Torso tilt

    return {
        'left_arm': left_arm_angle,
        'right_arm': right_arm_angle,
        'left_leg': left_leg_angle,
        'right_leg': right_leg_angle,
        'torso': torso_angle
    }

def detect_contact_frame(keypoints_df, frame_interval, reference_angles):
    """ Use the reference frame's angles (from keypoints_frame.csv) to detect the best-matching contact frame. """
    best_frame = None
    best_match_score = float('inf')  # Lower is better

    # Define weights for different angles
    angle_weights = {
        'left_arm': 4.0,    # Higher weight for arm angles
        'right_arm': 2.0,
        'left_leg': 4.0,    # Moderate weight for leg angles
        'right_leg': 2.0,
        'torso': 0.0        # Lower weight for torso angle
    }

    # Iterate through frames and calculate angles, compare to reference angles
    for frame in range(keypoints_df['frame'].min(), keypoints_df['frame'].max() + 1, frame_interval):
        frame_data = keypoints_df[keypoints_df['frame'] == frame]

        # Ensure all necessary keypoints exist for the frame
        if frame_data.empty or \
           frame_data[frame_data['body_part'] == 'left_wrist'].empty or \
           frame_data[frame_data['body_part'] == 'right_wrist'].empty:
            continue

        # Calculate angles for the current frame
        current_angles = get_pose_angles(frame_data)

        # Compare angles to reference and calculate the weighted difference
        total_diff = 0
        print(f"Frame {frame}:")  # Debugging print to show the frame number
        for key in reference_angles:
            angle_diff = abs(current_angles[key] - reference_angles[key])
            weighted_diff = angle_diff * angle_weights.get(key, 1.0)  # Default weight is 1.0 if not specified
            total_diff += weighted_diff
            print(f"  {key}: Angle Difference = {angle_diff:.2f}, Weighted Difference = {weighted_diff:.2f}")

        # If the match is closer, update the best frame
        if total_diff < best_match_score:
            best_match_score = total_diff
            best_frame = frame

        print(f"  Total weighted angle difference = {total_diff:.2f}\n")

    print(f"Best matching contact frame: {best_frame}")
    return best_frame




def calculate_extension_score(keypoints_df, contact_frame, frame_interval):
    # Initialize variables for extension calculation
    total_extension_score = 0.0
    num_frames = 0

    # Iterate through frames starting at the contact frame to calculate follow-through
    for frame in range(contact_frame, keypoints_df['frame'].max() + 1, frame_interval):
        frame_data = keypoints_df[keypoints_df['frame'] == frame]

        # Ensure all necessary keypoints exist for the frame
        if frame_data.empty or \
           frame_data[frame_data['body_part'] == 'left_shoulder'].empty or \
           frame_data[frame_data['body_part'] == 'left_elbow'].empty or \
           frame_data[frame_data['body_part'] == 'left_wrist'].empty or \
           frame_data[frame_data['body_part'] == 'right_shoulder'].empty or \
           frame_data[frame_data['body_part'] == 'right_elbow'].empty or \
           frame_data[frame_data['body_part'] == 'right_wrist'].empty:
            print(f"Frame {frame}: Missing arm keypoints. Skipping.")
            continue

        # Extract the shoulder, elbow, and wrist keypoints for both arms
        left_shoulder = frame_data[frame_data['body_part'] == 'left_shoulder'][['x', 'y']].values[0]
        left_elbow = frame_data[frame_data['body_part'] == 'left_elbow'][['x', 'y']].values[0]
        left_wrist = frame_data[frame_data['body_part'] == 'left_wrist'][['x', 'y']].values[0]
        right_shoulder = frame_data[frame_data['body_part'] == 'right_shoulder'][['x', 'y']].values[0]
        right_elbow = frame_data[frame_data['body_part'] == 'right_elbow'][['x', 'y']].values[0]
        right_wrist = frame_data[frame_data['body_part'] == 'right_wrist'][['x', 'y']].values[0]

        # Calculate arm extension angles (shoulder-elbow-wrist) using the law of cosines
        left_arm_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
        right_arm_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)

        # Measure the efficiency of arm extension: higher angle means better extension
        left_extension_score = (left_arm_angle / 180) * 100  # 100 for full extension
        right_extension_score = (right_arm_angle / 180) * 100
        avg_extension_score = (left_extension_score + right_extension_score) / 2

        total_extension_score += avg_extension_score
        num_frames += 1

        # Print the extension angle and score for each frame
        print(f"Frame {frame}: Left Arm Angle: {left_arm_angle:.2f}°, Right Arm Angle: {right_arm_angle:.2f}°")
        print(f"Frame {frame}: Average Extension Score: {avg_extension_score:.2f} / 100")

    # Calculate the final extension score as an average over the follow-through frames
    final_extension_score = total_extension_score / num_frames if num_frames > 0 else 0.0

    print(f"\nFinal Extension Score: {final_extension_score:.2f} / 100")
    return final_extension_score


# Example usage:
keypoints_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/keypoints.csv')
frame_interval = 3  # Use the specified frame interval

# Load the reference frame (keypoints_frame.csv)
reference_keypoints_df = pd.read_csv('/content/drive/MyDrive/HRnet/SwingSensei-AI-Model/Data/swing_pose_output/keypoints_frame.csv')

# Get the pose angles for the reference frame (keypoints_frame.csv)
reference_angles = get_pose_angles(reference_keypoints_df)

# Detect the contact frame using the reference keypoints data
contact_frame = detect_contact_frame(keypoints_df, frame_interval, reference_angles)

# Calculate the extension score starting from the detected contact frame
extension_score = calculate_extension_score(keypoints_df, contact_frame, frame_interval)
```
